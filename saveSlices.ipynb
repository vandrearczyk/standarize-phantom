```python
# To save the features:
array_slices = list()
array_labels = list()
array_cartridge = list()
array_scan = list()
for scan in range(17):
    # Load contour
    dcm1 = dicom.read_file(listFilesContour[scan]+"/000000.dcm")
    # Find paths to volumes
    PathDicom = listFilesVol[scan]
    lstFilesDCM = []  # create an empty list
    for dirName, subdirList, fileList in os.walk(PathDicom):
        for filename in fileList:
            if ".dcm" in filename.lower():  # check whether the file's DICOM
                lstFilesDCM.append(os.path.join(dirName,filename))
    # Get ref file
    RefDs = dicom.read_file(lstFilesDCM[0])

    # Load dimensions based on the number of rows, columns, and slices (along the Z axis)
    ConstPixelDims = (int(RefDs.Rows), int(RefDs.Columns), len(lstFilesDCM))

    # Load spacing values (in mm)
    ConstPixelSpacing = (float(RefDs.PixelSpacing[0]), float(RefDs.PixelSpacing[1]), float(RefDs.SliceThickness))

    # Load image position (in mm) for x and y (as z changes)
    ConstImagePosition = (float(RefDs.ImagePositionPatient[0]), float(RefDs.ImagePositionPatient[1]))
    
    print 'scan:',scan,'PixelSpacing:',ConstPixelSpacing, 'ConstImagePosition',ConstImagePosition
    
    x = np.arange(0.0, (ConstPixelDims[0]+1)*ConstPixelSpacing[0], ConstPixelSpacing[0])
    y = np.arange(0.0, (ConstPixelDims[1]+1)*ConstPixelSpacing[1], ConstPixelSpacing[1])
    z = np.arange(0.0, (ConstPixelDims[2]+1)*ConstPixelSpacing[2], ConstPixelSpacing[2])
    
    # The array is sized based on 'ConstPixelDims'
    ArrayDicom = np.zeros(ConstPixelDims, dtype=RefDs.pixel_array.dtype)
    ArrayImagePositions = np.zeros(len(lstFilesDCM), dtype=float)
    # loop across all the DICOM files
    for filenameDCM in lstFilesDCM:
        # read the file
        ds = dicom.read_file(filenameDCM)
        # store the raw image data
        ArrayDicom[:, :, lstFilesDCM.index(filenameDCM)] = ds.pixel_array
        ArrayImagePositions[lstFilesDCM.index(filenameDCM)] = ds.ImagePositionPatient[2]
    
    intercept = ds.RescaleIntercept
    slope = ds.RescaleSlope
    if slope != 1: # All slopes are actually 1
        ArrayDicom = slope * ArrayDicom.astype(np.float64)
        ArrayDicom = ArrayDicom.astype(RefDs.pixel_array.dtype)
    ArrayDicom = ArrayDicom + intercept

    # loop across cartridges
    for cartridge in range(len(dcm1.ROIContours)):
        print 'cartridge:', cartridge, 'nb slices:', len(dcm1.ROIContours[cartridge].Contours)
        # loop across slices in contour in this Cartridge
        i=0
        for sliceCart in range(len(dcm1.ROIContours[cartridge].Contours)):
        #for sliceCart in range(0,1):
            array_scan.append(scan)
            array_cartridge.append(cartridge)
            slicePos = dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData[2] # I need to get the index of this Slice Position in the volume as it is not sorted.
            
            # for this z, find (min x, max x) and (min y, max y)
            x = sorted((dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData[0],dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData[len(dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData)/2]))
            y = sorted((dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData[1],dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData[len(dcm1.ROIContours[cartridge].Contours[sliceCart].ContourData)/2+1]))

            # Convert to actual pixel position:
            xPix = (int((x[0] - ConstImagePosition[0])/ConstPixelSpacing[0]),int((x[1] - ConstImagePosition[0])/ConstPixelSpacing[0]))
            yPix = (int((y[0] - ConstImagePosition[1])/ConstPixelSpacing[1]),int((y[1] - ConstImagePosition[1])/ConstPixelSpacing[1]))

            # DENNIS SENT ME THE CORRECT ROIs I DON'T NEED IT ANYMORE ***************
            ## Somehow S2, T1, T2 and T3 are not alligned with the contours. I allign them with an offset
            #if scan == 13:
            #    offset = 472.8
            #elif scan == 14:
            #    offset = -824.5
            #elif scan == 15:
            #    offset = 734
            #elif scan == 16:
            #    offset = -789.5
            #else:
            #    offset = 0.0

            # I somehow mixed up x and y.
            x = ArrayDicom[yPix[0]:yPix[1],xPix[0]:xPix[1],np.where(np.around(ArrayImagePositions,decimals=2)==slicePos)[0][0]]# for z, I do an uggly index from the dicom volume position. There is for sure a better way.  
            array_slices.append(x)
            
            i+=1
            if scan in (0,6,13,14,16):
                array_labels.append(9-cartridge)
            else: 
                array_labels.append(cartridge)
```
